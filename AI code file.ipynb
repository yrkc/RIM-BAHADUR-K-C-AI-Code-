```python
#Import The Necessary Library
import math
import ipywidgets as widgets
from IPython.display import display, clear_output
from sklearn.tree import DecisionTreeClassifier
import numpy as np
```

```python
#Take Manual Samle Data For Implementing The Concept Of AI
Student_Information = [
    [5, 8, 1, 0],  # Correct: Triangle for a garden
    [4, 7, 0, 0],  # Incorrect: Triangle for a garden
    [7, 3, 1, 1],  # Correct: Triangle for wall painting
    [5, 2, 0, 1],  # Incorrect: Triangle for wall painting
    [6, 9, 0, 2],  # Incorrect: Parallelogram for construction
    [3, 4, 1, 2],  # Correct: Rectangle for construction
    [4, 4, 0, 1],  # Incorrect: Square for wall painting
]
```

```python
# Training Process For Decision Tree
Decision_X = np.array([i[:2] for i in Student_Information])
Decision_y = np.array([i[2] for i in Student_Information])
clf = DecisionTreeClassifier()
clf.fit(Decision_X, Decision_y)
```

```python

def Area_Hub(S, B, H=None, R=None):

    if S == "Triangle":
        return 0.5 * B * H
    elif S == "Rectangle":
        return B * H
    elif S == "Square":
        return B ** 2
    elif S == "Circle":
        return math.pi * R ** 2
    elif S == "Parallelogram":
        return B * H
    else:
        return None
```

```python
# Implementation Of Ai Function
def Prediction_Zone(B, H=None, R=None, S=None):

    #  Feature List For Prediction
    if H is not None:
        feature = [B, H]  #  For Triangle
    else:
        feature = [B, H]  # For circle base will be used as Radious

    prediction = clf.predict([feature])[0]

    if prediction == 1:
        feedback = "Good job! Keep it up."
    else:
        feedback = "It looks like you need a little more practice. Try again!"

    # summed up some personized schenerio for better visualization
    if S == 0:
        feedback += "\n It will be helpfull to plan a Garden."
    elif S == 1:
        feedback += "\nFor estimating the paint on the Wall it will be helpfull"
    elif S == 2:
        feedback += "\nDuring construction project time for calculating the correct area calculatiuon it will be helpfull for material estimation purpose"

    return feedback
```

```python

# Attractive UI design
Shape_Store = widgets.Dropdown(
    options=["Triangle", "Rectangle", "Square", "Circle", "Parallelogram"],
    value="Triangle",
    description="Shape:"
)

Base_Store = widgets.FloatText(
    value=0,
    description="Base:"
)

Height_Store = widgets.FloatText(
    value=0,
    description="Height:"
)

Radious_Store = widgets.FloatText(
    value=0,
    description="Radius:"
)

Schenerio_Store = widgets.Dropdown(
    options=["Garden", "Wall Painting", "Construction Project"],
    value="Garden",
    description="Scenario:"
)

output_area = widgets.Output()
```

```python
def Visibility_After_Click(_):

    SHAPE = Shape_Store.value
    SCHENERIO = Schenerio_Store.index

    try:
        BASE = Base_Store.value
        HEIGHT = Height_Store.value if Height_Store.value else None
        RADIOUS = Radious_Store.value if Radious_Store.value else None

        area = Area_Hub(SHAPE, BASE, HEIGHT, RADIOUS)
        with output_area:
            clear_output()
            if area is not None:
                print(f"The area of the {SHAPE} is {area:.2f} square units.")
                feedback = Prediction_Zone(BASE, HEIGHT, RADIOUS, SCHENERIO)
                print(feedback)
            else:
                print("INVALID INPUT.")
    except ValueError:
        with output_area:
            clear_output()
            print("PLEASE ENTER THE VALID VALUE")
```

```python
# Calculate Button
calculate_button = widgets.Button(
    description="Calculate Area",
    button_style="success"
)
calculate_button.on_click(Visibility_After_Click)
```

```python
# Arrange Widgets
ui = widgets.VBox([
    Shape_Store,
    Base_Store,
    Height_Store,
    Radious_Store,
    Schenerio_Store,
    calculate_button,
    output_area
])
```

```python
# Display the Interface
display(ui)
```
